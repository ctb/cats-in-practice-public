# 15genome - search strategies comparison

Here we are taking 15 complete genomes, loading them into a graph, and building a catlas. We are then searching with MinHashes of each of those genomes using several strategies, and comparing the nodes we get back with ground truth.

Note, this uses a catlas with radius 3.

```python
fjacc = numpy.loadtxt('15genome-catlas-fjacc.csv',
                        delimiter=',', skiprows=1, usecols=(0,1))
fworst = numpy.loadtxt('15genome-catlas-fworst.csv',
                        delimiter=',', skiprows=1, usecols=(0,1))
gathermins = numpy.loadtxt('15genome-catlas-gathermins.csv',
                        delimiter=',', skiprows=1, usecols=(0,1))
```

```python

fig, axes = pylab.subplots(nrows=1, ncols=3, figsize=(8, 5))

def plot_pair(what, axes, first, name):
    axes[first].violinplot(what, showmedians=True)
    axes[first].axis(ymin=0, ymax=100)
    #axes[first].get_xaxis().set_visible(False)
    axes[first].set_title(name)
    axes[first].set_xticks([1, 2])
    axes[first].set_xticklabels(['sensitivity', 'specificity'])
    

plot_pair(fjacc, axes, 0, 'fjacc')
plot_pair(fworst, axes, 1, 'fworst')
plot_pair(gathermins, axes, 2, 'gathermins')
```
